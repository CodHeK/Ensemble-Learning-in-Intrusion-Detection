```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
features = pd.read_csv("features.txt",sep=",")
```

```python
df = pd.read_csv("kdd10.csv", names = features.columns)
df.head()
```

```python
df = df.reset_index()
```

```python
df['attack_type'] = df["attack_type"].astype('category')
df['attack_type'] = df['attack_type'].cat.codes
df['attack_type'] = df["attack_type"].astype('int64')
```

```python
dummies = pd.get_dummies(df['protocol_type'], drop_first='True')
df = pd.concat([df, dummies], axis=1)
df.drop(['protocol_type'], inplace=True, axis=1)

dummies = pd.get_dummies(df['service'], drop_first='True')
df = pd.concat([df, dummies], axis=1)
df.drop(['service'], inplace=True, axis=1)

dummies = pd.get_dummies(df['flag'], drop_first='True')
df = pd.concat([df, dummies], axis=1)
df.drop(['flag'], inplace=True, axis=1)
df.head()
```

```python
X = df.drop(columns=['attack_type'],axis=1)
y = df['attack_type']

X = X.fillna(X.mean())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
glf = GaussianNB()
glf.fit(X_train, y_train)
```

```python
pred = glf.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix

import sys
sys.path.append('/usr/local/lib/python2.7/dist-packages')
import seaborn as sns
```

```python
# sns.heatmap(confusion_matrix(pred,y_test),cmap='coolwarm')
print(classification_report(pred,y_test))
```
